```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Skyserver_SQL2_27_2018 6_51_39 PM.csv')
data
```

```python
g = data["g"]
g_r = data["g"].sub(data["r"])

g = g.values
g_r = g_r.values
```

```python
import matplotlib.pyplot as plt

plt.hist2d(g_r,g, bins = 100)
plt.xlabel("g-r",fontsize=20)
plt.ylabel("g",fontsize=20)
plt.xlim(-0.5,1.5)
plt.show()
```
